```python
def FizzBuzz(i):
    if i%3 == 0 and i%5 == 0:
        return "FizzBuzz"
        #3で割り切れて、なおかつ5で割り切れる数のときは "FizzBuzz"
    if i %3 == 0:
        return "Fizz"
        #3で割り切れる数のときは "Fizz"
    if i %5 == 0:
        return "Buzz"
        #5で割り切れる数のときは "Buzz"
    else:
        return str(i)
        #それ以外の数値の場合は、その数値を文字列に変換したデータ    

s = [FizzBuzz(i) for i in range(1,101)]

s
```
